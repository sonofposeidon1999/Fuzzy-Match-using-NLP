```python
import pandas as pd
from nltk.tokenize import RegexpTokenizer
import nltk
from fuzzywuzzy import fuzz

ln=pd.read_csv('/Users/daksha.singhal/Desktop/Daksh Records.csv',encoding = "ISO-8859-1", low_memory=False)

for i in range(len(ln['Ledger Narration 1'])):
    if(pd.isna(ln['Ledger Narration 1'][i])):
        ln['Ledger Narration 1'][i]=''
for i in range(len(ln['Narration 1'])):
    if(pd.isna(ln['Narration 1'][i])):
        ln['Narration 1'][i]=''

ln['Narration 1']=ln['Narration 1'].astype('str')
ln['Ledger Narration 1']=ln['Ledger Narration 1'].astype('str')
ln['Ledger Narration 2']=ln['Ledger Narration 2'].astype('str')
ln['Ledger Narration 3']=ln['Ledger Narration 3'].astype('str')

unq_col=list(ln['Ledger Narration 1'].unique())
unq_words1=[]
for i in unq_col:
    for j in i.split():
        unq_words1.append(j)
unq_col=list(ln['Ledger Narration 2'].unique())
unq_words2=[]
for i in unq_col:
    for j in i.split():
        unq_words2.append(j)
unq_col=list(ln['Ledger Narration 3'].unique())
unq_words3=[]
for i in unq_col:
    for j in i.split():
        unq_words3.append(j)

unq_words=set(unq_words1+unq_words2+unq_words3)

stop=[i for i in stop if len(i)>1]
ln['Ledger Narration 1'] =ln['Ledger Narration 1'].apply(lambda x: " ".join(x.lower() for x in x.split())) 
ln['Ledger Narration 1']=ln['Ledger Narration 1'].str.replace('[^\w\s]',' ')
freq = pd.Series(' '.join(ln['Ledger Narration 1']).split()).value_counts()[:5] ##common words removal
freq = list(freq.index)
ln['Ledger Narration 1'] =ln['Ledger Narration 1'].apply(lambda x: " ".join(x for x in x.split() if x not in (stop+freq)))
ln['Ledger Narration 1'] = ln['Ledger Narration 1'].apply(lambda x: " ".join(x for x in x.split() if x.isalpha()==True))

ln['Ledger Narration 2'] =ln['Ledger Narration 2'].apply(lambda x: " ".join(x.lower() for x in x.split())) 
ln['Ledger Narration 2']=ln['Ledger Narration 2'].str.replace('[^\w\s]',' ')
freq = pd.Series(' '.join(ln['Ledger Narration 2']).split()).value_counts()[:5] ##common words removal
freq = list(freq.index)
ln['Ledger Narration 2'] =ln['Ledger Narration 2'].apply(lambda x: " ".join(x for x in x.split() if x not in (stop+freq)))
ln['Ledger Narration 2'] = ln['Ledger Narration 2'].apply(lambda x: " ".join(x for x in x.split() if x.isalpha()==True))

ln['Ledger Narration 3'] =ln['Ledger Narration 3'].apply(lambda x: " ".join(x.lower() for x in x.split())) 
ln['Ledger Narration 3']=ln['Ledger Narration 3'].str.replace('[^\w\s]',' ')
freq = pd.Series(' '.join(ln['Ledger Narration 3']).split()).value_counts()[:5] ##common words removal
freq = list(freq.index)
ln['Ledger Narration 3'] =ln['Ledger Narration 3'].apply(lambda x: " ".join(x for x in x.split() if x not in (stop+freq)))
ln['Ledger Narration 3'] = ln['Ledger Narration 3'].apply(lambda x: " ".join(x for x in x.split() if x.isalpha()==True))

ln['Narration 1'] =ln['Narration 1'].apply(lambda x: " ".join(x.lower() for x in x.split())) 
ln['Narration 1']=ln['Narration 1'].str.replace('[^\w\s]',' ')
freq = list(set(pd.Series(' '.join(ln['Narration 1']).split()).value_counts()[:50].index)-unq_words) ##common words removal
ln['Narration 1'] = ln['Narration 1'].apply(lambda x: " ".join(x for x in x.split() if x not in (stop+freq)))
ln['Narration 1'] = ln['Narration 1'].apply(lambda x: " ".join(x for x in x.split() if x.isalpha()==True))

ln['score1'] = ln.apply(lambda x:fuzz.token_sort_ratio( x['Narration 1'] , x['Ledger Narration 1']), axis=1)
ln['score2'] = ln.apply(lambda x:fuzz.token_sort_ratio( x['Narration 1'] , x['Ledger Narration 2']), axis=1)
ln['score3'] = ln.apply(lambda x:fuzz.token_sort_ratio( x['Narration 1'] , x['Ledger Narration 3']), axis=1)

ln=ln[['Narration 1','Ledger Narration 1','score1','Ledger Narration 2','score2','Ledger Narration 3','score3']]

ln.to_csv('/Users/daksha.singhal/Desktop/result8.csv', encoding='utf-8', index=False)

pd.Series(' '.join(ln['Narration 1']).split()).value_counts()[:50]

```
